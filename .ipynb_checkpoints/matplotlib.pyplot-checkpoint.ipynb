# matplotlib.pyplot 모듈의 간단한 예제

```python
import matplotlib.pyplot as plt
```

### PLOT

```python
plt.plot([1, 2, 3, 4]) 
plt.show()
```

> plot함수는 리스트의 값들이 y값들이라고 가정하고
x값 [0,1,2,3]을 자동으로 만들어 낸다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

> plot() 함수는 다양한 기능을 포함하고 있어서, 임의의 개수의 인자를 받을 수 있다.
예를 들어 위와 같이 입력하면 x-y 값을 그래프로 나타낼 수 있다.

### 스타일도 지정 가능함.
### 'ro' -> 빨간색의 원형(o)마커.
### 'b-' -> 파란색의 실선(-)


```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

> axis->[xmin, xmax, ymin, ymax]
